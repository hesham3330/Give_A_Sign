```python
from ultralytics import YOLO
import os
# change directory
os.chdir(r"E:\projects\Give-A-Sign\English")
```

```python
# A function used to reduce the number of video frames for fast model performance
from moviepy.editor import VideoFileClip
lenght= 0
def get_video_length(video_path):
    clip = VideoFileClip(video_path)
    duration_seconds = clip.duration
    duration_seconds %= 60
    clip.close()
    return duration_seconds
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
input_path = 'E://projects//Give-A-Sign//English//test1.jpg'
# Read and display the image
img = mpimg.imread('E://projects//Give-A-Sign//English//test1.jpg')
imgplot = plt.imshow(img)
plt.axis('off') 
plt.show() 
lenght = get_video_length(input_path)
length= int(lenght*2)
```

```python
# the model import
model = YOLO('best.pt')
model.predict(source = input_path,conf=0.5, save = True,vid_stride = length , save_crop = True,project='E://projects//Give-A-Sign//English', name='done')

# function to detremin if there is a result     
def folder_exists(folder_path):
    return os.path.exists(folder_path)
```

```python
# Example usage
list_of_labels = []
folder_path = 'E://projects//Give-A-Sign//English//done/crops'
if folder_exists(folder_path):
    # get output of prdedet from this pass
    for folder in os.listdir('E://projects//Give-A-Sign//English//done/crops'):
        list_of_labels.append(folder)
else:
    list_of_labels.append("NO DETECTION")


# print the items of the list of outputs
for i in list_of_labels:
    print(i)
```
