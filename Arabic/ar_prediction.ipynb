```python
# Function to get label
list_of_labels = []
label= "B"
def get_label(label):
    for key , value in labels.items():
        if (key == label):
            label = value
    return label
```

```python
# A function used to reduce the number of video frames for fast model performance
from moviepy.editor import VideoFileClip
lenght= 0
def get_video_length(video_path):
    clip = VideoFileClip(video_path)
    duration_seconds = clip.duration
    duration_seconds %= 60
    clip.close()
    return duration_seconds
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
input_path = 'E://projects//Give_A_Sign//Arabic//test1.jpeg'
# Read and display the image
img = mpimg.imread('E://projects//Give_A_Sign//Arabic//test1.jpeg')
imgplot = plt.imshow(img)
plt.axis('off') 
plt.show() 
lenght = get_video_length(input_path)
length= int(lenght*2)
```
